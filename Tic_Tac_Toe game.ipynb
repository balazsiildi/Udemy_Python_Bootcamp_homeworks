```python
#To clear the output we import the necessary function, random is needed to alternate between players who starts
from IPython.display import clear_output
import random

def start_game(is_first_game):
    '''
    Explaining the game to the players and asking for the choice of character the first player want to use.
    Parameters: is_first_game boolean, if true, rules are explained
    Returns: is_first_game=False, 
             first_goes= 1 or 2, who starts the game
             play1_char="X" or "O"
    '''
    #Initialazing the chosen value to be non-acceptable
    play1_char="WRONG"

    #Clear the output before starting the game
    clear_output()
    
    #If this is the first game in this run, explain the rules
    if is_first_game:
        
        print("This is a tic-tac-toe game for two players")
        print("You indicate your step by chosing a number between 1 to 9 according to the following board arrangement:")
        display_board(board_buttons)

    #setting is_first_game to false
    is_first_game=False

    #deciding randomly, who goes first
    if random.randint(0,1)==1:
        first_goes=1
    else:
        first_goes=2

    print(f"Player {first_goes}, you start the game.")
    
    #Choosing the character the first player plays with, with checking that they gave a correct character(non-capital letters are also accepted)
    while play1_char not in ["X","O"]:
    
        play1_char=input(f"Player {first_goes}, which character do you choose to play with (X or O)? ").upper()

        if play1_char not in ["X","O"]:
            print("Sorry, your choice is not valid.")
    
    return is_first_game, first_goes, play1_char

def display_board(board):
    '''
    Displays the board at its current status
    Input: board - the dictionary in which we store the board.
    '''
    print(fill_row)
    print(board["row1"])
    print(fill_row)
    print(board_sep_row)
    print(fill_row)
    print(board["row2"])
    print(fill_row)
    print(board_sep_row)
    print(fill_row)
    print(board["row3"])
    print(fill_row)

def chosing_step(step, first_goes, board):
    '''
    Asks the current player (based on step and who started) to choose a position. 
    Checks if the given input looks as expected and if the position is still free and asks a new input until it is.
    Inputs: step int - which step are we in, 
            first_goes= 1 or 2, who starts the game
            board dict - the current status of the board
    Outputs: row str - indicats which row the player choose, 
             place int - indicates which character to manipulate in the choosen row
    '''
    #Initialazing the variables
    position="WRONG"
    position_taken=True
    acceptable_choices=['1','2','3','4','5','6','7','8','9']
    
    #Assigning the player based on step
    if (step % 2 == 0 and first_goes==1) or (step % 2 == 1 and first_goes==2) :
        player="Player2"
    else:
        player="Player1"
    
    #Choosing the position with checking that they gave a correct digit and whether the chosen place is still free
    while position not in acceptable_choices or position.isdigit()==False or position_taken==True:
       
        position=input(f"{player}, which position would you like to take? Choose a number between 1 to 9: ")

        if position.isdigit()==False:
            print("Sorry, your choice is not a digit!")
        else:
            if int(position) not in range(1,10):
                print("Sorry, your choice is out of range!")
            else:
                # given the position, the row and place we should manipulate can be calculated
                if board["row"+str((int(position)-1) // 3 + 1)][(int(position)-1) % 3 * 6 + 2] == ' ': 
                    position_taken=False
                if position_taken:
                    print("Sorry, that position is no longer available, please choose a position not taken before.")
    
    return "row"+str((int(position)-1)//3+1), (int(position)-1)%3*6 + 2

def rewrite_board(step, play1_char, board, row, place):
    '''
    Writes the correct sign on the board. 
    Inputs: step int - in which step are we in, 
            play1_char str - the character the first player choose to play with
            board dict - the status of the board
            row str - the row the last player choose
            place int - the place we should manipulate in the choosen row
    Output: the new status of the board
    '''
    #Calculating what character to write in the choosen row and place based on the step and the character choosen by the first player
    if (step % 2 == 1 and play1_char == "O") or (step % 2 == 0 and play1_char == "X"):
        new_sign="O"
    else:
        new_sign="X"

    #Writing the correct character into the choosen place of the board 
    board[row]=board[row][:place]+new_sign+board[row][place+1:]

    #Clearing the output
    clear_output()

    #Displaying the new board
    print("Here is the current board: ")
    display_board(board)

    return board

def winning_board(board, row, place):
    '''
    Checks whether with the last step, we have a winner. 
    Inputs: board dict - the status of the board
            row str - the row the last player choose
            place int - the place last player choose
    Output: boolean
    '''

    #Checking whether the last manipulated row of the board has identical characters in the three positions
    if board[row][2]==board[row][8]==board[row][14]:
        return True
        
    #Checking whether the last manipulated position of the board has identical characters in the three rows
    if board["row1"][place]==board["row2"][place]==board["row3"][place]:
        return True

    #Checking diagonals of the board if the last choosen position is in the corners or in the middle
    if (row=="row2" and place==8) or (row in ["row1","row3"] and place in (2,14)):
        if (board["row1"][2]==board["row2"][8]==board["row3"][14]!=' ') or (board["row1"][14]==board["row2"][8]==board["row3"][2]!=' '):
            return True

    #If non of the above are true, we don't have a winner yet
    return False

def closing_the_game(step,first_goes,winner):
    '''
    At the end of the game congratulates the winner (if we have one) and asks the players whether they want to play more. 
    Inputs: step int - in which step the game ended, to know who won
            first_goes= 1 or 2, who starts the game
            winner boolean - whether we have a board with three indentical signs inline.
    Output: boolean
    '''
    #inititalizing the indicator
    play_indicator="WRONG"

    #If we have a winner, congratulate him/her
    if winner==True:
        
        #Assigning the player who was the last to choose a position based on step
        if (step % 2 == 0 and first_goes==1) or (step % 2 == 1 and first_goes==2):
            player="Player1"
            wins[0] += 1
        else:
            player="Player2"
            wins[1] += 1
            
        print(f"Congratulations, {player}! You won the game! ")
    else:
        print("Non of you could beat the other! Maybe next time!")
        wins[2] += 1
    
    #Asking if they want to play a new game, with checking that they gave a correct character(non-capital letters are also accepted):
    while play_indicator not in ["Y","N"]:
    
        play_indicator=input("Would you like to play again (Y or N)? ").upper()

        if play_indicator not in ["Y","N"]:
            print("Sorry, your choice is not valid.")

    #Saying goodbye
    if play_indicator=="N":
        clear_output()
        print("Thanks for the game! It was my pleasure!")
        print(f"You played {wins[0]+wins[1]+wins[2]} times, Player 1 won {wins[0]} game(s), Player 2 won {wins[1]}.")

    
    return play_indicator=="Y"
```

```python
#Initializing the constants
is_first_game=True
fill_row="     |     |     "
board_sep_row="-----------------"
board_buttons={"row1": "  1  |  2  |  3  ", "row2": "  4  |  5  |  6  ", "row3": "  7  |  8  |  9  "}
wins=[0,0,0]

want_to_play=True

#While want to play
while want_to_play==True:

    
    #Starting the game and explaining the rules if it is the first time
    is_first_game, first_goes, play1_char=start_game(is_first_game)

    #Initializong step, clearing the board and setting winner to false
    step=1
    board_value={"row1": "     |     |     ", "row2": "     |     |     ", "row3": "     |     |     "}
    winner=False

    #While there is no winner and there are still free positions on the board
    while winner==False and step<10:

        #Asking the player to choose a position
        row,place=chosing_step(step, first_goes, board_value)

        #Rewriting the board and displaying it's new status
        rewrite_board(step, play1_char, board_value, row, place)

        #Checking if there is a winner
        winner=winning_board(board_value, row, place)

        #Increasing the step
        step += 1

    #Close the game with reacting to the result and asking for a new round
    want_to_play=closing_the_game(step,first_goes,winner)
```
